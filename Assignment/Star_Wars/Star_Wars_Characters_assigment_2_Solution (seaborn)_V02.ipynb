```python
df["mass_cat"] = df.mass.apply(lambda x : "High_mass" if x >=100.0 else  ("Normal_mass" if x> 50.0 and x<100.0 else "Low_mass" if x<=50  else np.nan ))
```

```python
df["mass_cat"].value_counts(dropna= False)
```

```python
mass_cat = df["mass_cat"].value_counts()
```

```python
# with seaborn barplot

fig, ax = plt.subplots(figsize=(8, 5))

ax = sns.barplot(x=mass_cat.index, y=mass_cat.values)

ax.set_xticklabels(mass_cat.index, rotation=0)
ax.set_title('Mass Groups', color='blue', fontsize=15)
for container in ax.containers:
    ax.bar_label(container);
```

```python
# with seaborn barplot_2

fig, ax = plt.subplots(figsize=(8, 5))

ax = sns.barplot(x=df["mass_cat"].value_counts().index, y=df["mass_cat"].value_counts().values)

ax.set_xticklabels(df["mass_cat"].value_counts().index, rotation=0)
ax.set_title('Mass Groups', color='blue', fontsize=15)
for container in ax.containers:
    ax.bar_label(container);
```

```python
# with seaborn countplot
palet= sns.color_palette("rocket_r", n_colors=5) 

fig, ax = plt.subplots(figsize=(8, 5))

sns.countplot(x="mass_cat",
              data=df,
              order=df["mass_cat"].value_counts().index,
              palette = palet)

ax.set_xticklabels(df["mass_cat"].value_counts().index, rotation=45)
ax.set_title('Mass Groups', color='blue', fontsize=15);
ax.set(facecolor = "powderblue")
fig.set_facecolor('white')
ax.plot(0, 33, "*", markersize=10, color="r"); # x=0 y= 33 --> normal mass 33 values point

for p in ax.patches:
    ax.annotate(round(p.get_height()), (p.get_x()+0.35, p.get_height()+0.5), color="red");
```
